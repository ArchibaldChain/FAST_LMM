Let the model be
$$
Y_i = \sum_{j = 1}^{p} \beta_j X_{i,j} + u_i +\epsilon_i \ \text{where } i = 1, 2,3,\dots,n
$$
where $(u_1, u_2, \dots, u_n)^\top \sim MVN(0, \sigma_g^2 K ) $ and $\epsilon_i \sim ^{i.i.d} N(0, \sigma_e^2)$, and  $K = 1/s_c W W^\top$, $\sigma_e^2 = \frac{1 - h^2}{h^2} \sigma_g^2$ . 

```python
SNP_ = SNP.copy()
SNP_.drop('Y', inplace= True, axis = 1)
data = SNP_.values
y = SNP['Y'].values
X = data[:,:6]
W = data[:, 1:]
```
